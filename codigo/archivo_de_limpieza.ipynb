```python
# TITULO: MODELO PARA LA EVALUACIÓN DE FOTOMULTAS EN CDMX
# AUTOR: EDUARDO HERNANDEZ MARTINEZ
# FECHA DE ULTIMA MODIFICACION: 05/05/2026
#----------------------------------------------- DECLARACIÓN DE LIBRERÍAS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from geopy.distance import geodesic
#----------------------------------- CARGA DE LOS DATOS
df1 = pd.read_csv('inViales_2014_2015.csv', on_bad_lines='skip')
df2 = pd.read_csv('inViales_2016_2018.csv', on_bad_lines='skip')
df3 = pd.read_csv('inViales_2019_2021.csv', on_bad_lines='skip')
df4 = pd.read_csv('inViales_2022_2024.csv', on_bad_lines='skip')

fotomultas = pd.read_csv('fotomultas.csv')
```

Renombramos las columnas reducirlas y normalizarlas

```python
df1 = df1.rename(columns={
    'fecha_creacion': 'fecha',
    'hora_creacion': 'hora',
    'tipo_incidente_c4': 'categoria',
    'incidente_c4': 'secuelas'
})
```

```python
df2 = df2.rename(columns={
    'fecha_creacion': 'fecha',
    'hora_creacion': 'hora',
    'tipo_incidente_c4': 'categoria',
    'incidente_c4': 'secuelas'
})
```

```python
df3 = df3.rename(columns={
    'fecha_creacion': 'fecha',
    'hora_creacion': 'hora',
    'tipo_incidente_c4': 'categoria',
    'incidente_c4': 'secuelas'
})
```

```python
df4 = df4.rename(columns={
    'fecha_creacion': 'fecha',
    'hora_creacion': 'hora',
    'tipo_incidente_c4': 'categoria',
    'incidente_c4': 'secuelas'
})
```

Se estableccen las columnas que nos aportan información para nuestro código como columnas importantes

```python
columnas_importantes= ['fecha', 'hora', 'categoria','secuelas', 'alcaldia_inicio' ,'latitud', 'longitud']
df1=df1[columnas_importantes]
df2=df2[columnas_importantes]
df3=df3[columnas_importantes]
df4=df4[columnas_importantes]
```

Normalizamos la columna de fecha en las 4 tablas

```python
df1['fecha'] = pd.to_datetime(df1['fecha'], errors='coerce')
df2['fecha'] = pd.to_datetime(df2['fecha'], errors='coerce')
df3['fecha'] = pd.to_datetime(df3['fecha'], errors='coerce')
df4['fecha'] = pd.to_datetime(df4['fecha'], errors='coerce')
```

```python
print(df2['latitud'].isna().sum())
print((df2['latitud']==0).sum())
print(df2['longitud'].isna().sum())
print((df2['longitud']==0).sum())
```

```python
print(df3['latitud'].isna().sum())
print((df3['latitud']==0).sum())
print(df3['longitud'].isna().sum())
print((df3['longitud']==0).sum())
```

Al solo contar con el 2do dataset con valores vacíos en la ubicación entonces se eliminan esos datos ya que son una cantidad minima

```python
df2 = df2[(df2['latitud'] != 0) & (df2['longitud'] != 0)]
df2 = df2.dropna(subset=['latitud', 'longitud'])
```

Obtenemos el resultado de camaras con datos faltantes

```python
total = len(fotomultas)
# vacios en cualquiera de las dos
na_total = fotomultas[['latitud', 'longitud']].isna().any(axis=1).sum()

print(f"Registros vacios en latitud o longitud: {na_total} ({na_total/total*100:.2f}%)")
```

Debido a que representa menos de un 20% y no podemos tener la capacidad de predecir la ubicación de las demás, optamos por eliminar aquellos registros con ubicacion nula

```python
fotomultas_limpio = fotomultas.dropna(subset=['latitud', 'longitud']).copy()

print(f"Registros originales: {len(fotomultas)}")
print(f"Registros limpios: {len(fotomultas_limpio)}")
```

```python
print(fotomultas_limpio['latitud'].isna().sum())
print((fotomultas_limpio['latitud']==0).sum())
print(fotomultas_limpio['longitud'].isna().sum())
print((fotomultas_limpio['longitud']==0).sum())
```
